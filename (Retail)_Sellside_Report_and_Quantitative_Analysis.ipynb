1. Sellside Analyst 리포트에서 정성적인 내용을 파악

```python
import os
import shutil
import requests
import pandas as pd
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
from tqdm import tqdm
from PyPDF2 import PdfReader
import openai
import json
from concurrent.futures import ThreadPoolExecutor, as_completed

# OpenAI API 키 설정
openai_key = "...."
gpt_model = "gpt-3.5-turbo"

WM_TOP_STOCKS = ['003670.KS', '051910.KS', '005930.KS', '011200.KS', '450080.KS', '015760.KS',  '003410.KS', '068270.KS', '005490.KS', '042700.KS', '068760.KS', '035420.KS']

WM_KOREA_STOCKS = [code.split(".")[0] for code in WM_TOP_STOCKS]
REPORT_FOLDER = "wm_korea_stocks_reports"

if not os.path.exists(REPORT_FOLDER):
    os.makedirs(REPORT_FOLDER)

def fetch_stock_data(item_code):
    url = f'https://finance.naver.com/research/company_list.naver?searchType=itemCode&itemCode={item_code}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    ul_element = soup.find('table', class_='type_1')
    if not ul_element:
        return None

    rows = []
    for tr in ul_element.find_all('tr'):
        row = []
        for td in tr.find_all('td'):
            if td.find('a'):
                row.append(td.find('a')['href'])
            else:
                row.append(td.get_text().strip())
        if len(row) == 6:
            try:
                row[4] = f"20{row[4][:2]}-{row[4][3:5]}-{row[4][6:8]}"
                row[5] = int(row[5])
                rows.append(row)
            except (IndexError, ValueError):
                continue

    df = pd.DataFrame(rows, columns=['항목', '내용', '출처', '파일 링크', '날짜', '조회수'])
    df['날짜'] = pd.to_datetime(df['날짜'])
    one_month_ago = datetime.now() - timedelta(days=31*2)
    recent_df = df[df['날짜'] >= one_month_ago].drop_duplicates().reset_index(drop=True)
    return recent_df

def download_file(url, folder_path, date):
    try:
        file_name = date.strftime("%Y%m%d_") + url.split('/')[-1]
        file_path = os.path.join(folder_path, file_name)
        response = requests.get(url)
        with open(file_path, 'wb') as file:
            file.write(response.content)
        return file_path
    except Exception as e:
        print(f"Error downloading file: {e}")
        return None

def summarize_pdf(file_path, item_code, api_key):
    try:
        with open(file_path, "rb") as file:
            reader = PdfReader(file)
            pdf_text = "".join([page.extract_text() for page in reader.pages])

        summary = get_intelligence(item_code, pdf_text, api_key)
        return summary
    except Exception as e:
        print(f"Error summarizing PDF: {e}")
        return None

def get_intelligence(ticker, text, api_key):
    prompt = f"""
투자 전문가로서 {ticker}에 대한 <정보>를 토대로
- 투자 기회와 위험에 대해서 작성.
- 새로운 제품 출시 또는 컨퍼런스 등과 같이 향후 주가에 영향을 줄 수 있는 이벤트를 정리.
- <정보>에 기반한 내용 외에는 작성하지 않음. 의견을 제시하지 않음.
- <Writing Rule>에 따라 <양식>에 맞춰 한글로 작성.

<Writing Rule>
1. 과도한 현재진행형 표현을 삼가해줘.
2. 명사형 표현보다는 동사형 표현을 서술.
3. 대명사 사용을 줄여.
4. 피동문 표현보다는 능동문 표현을 사용.
5. 무격조사 '의'를 과도하게 사용하는 것을 자제.
6. 경어체 표현을 적절히 활용.

<정보>
{text}

<양식>
##{ticker}에 대한 투자 기회, 위험 요인 그리고 이벤트

1. 투자 기회
   -
   -

2. 위험 요인
   -
   -

3. 향후 이벤트
   -
   -

(끝)
"""

    return openai_call(prompt, api_key)

def openai_call(instruction, api_key):
    try:
        api_url = "https://api.openai.com/v1/chat/completions"
        headers = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"}
        data = {"model": gpt_model, "messages": [{"role": "user", "content": instruction}], "top_p": 1.0}
        response = requests.post(api_url, headers=headers, data=json.dumps(data))
        return json.loads(response.text)["choices"][0]["message"]["content"].strip()
    except requests.exceptions.Timeout:
        return "요청 시간 초과"
    except Exception as e:
        return f"오류 발생: {e}"

def process_stock(item_code):
    folder_path = os.path.join(REPORT_FOLDER, item_code)
    if os.path.exists(folder_path):
        shutil.rmtree(folder_path)
    os.makedirs(folder_path, exist_ok=True)

    df = fetch_stock_data(item_code)
    if df is None:
        return f"No data for stock {item_code}"

    file_paths = []
    with ThreadPoolExecutor() as executor:
        future_to_url = {executor.submit(download_file, row['파일 링크'], folder_path, row['날짜']): row for _, row in df.iterrows()}
        for future in as_completed(future_to_url):
            result = future.result()
            if result:
                file_paths.append(result)

    summary_results = []
    for file_path in file_paths:
        summary = summarize_pdf(file_path, item_code, openai_key)
        if summary:
            summary_results.append(summary)

    if len(summary_results) > 0:
        print(f"\nSummary for stock {item_code}:\n")
        for summary in summary_results:
            print(summary)
            print("\n" + "=" * 50 + "\n")

# 주식 코드 처리
#for item_code in tqdm(WM_KOREA_STOCKS, desc="Processing stocks"):
for item_code in WM_KOREA_STOCKS:
    process_stock(item_code)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.relativedelta import relativedelta
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
import plotly.graph_objects as go

# Fetch stock data from Yahoo Finance
def get_stock_performance(ticker):

    #if "." in ticker:
    #    ticker=ticker.split(".")[0]

    stock = yf.Ticker(ticker)
    df = stock.history(period="1y")
    df.index=[str(x).split()[0] for x in list(df.index)]
    df.index.rename("Date",inplace=True)

    df['1Wk'] = df['Close'].pct_change(5)
    df['1M'] = df['Close'].pct_change(20)
    df['3M'] = df['Close'].pct_change(60)
    df['6M'] = df['Close'].pct_change(120)
    df['1Yr'] = df['Close'].pct_change(len(df)-1)

    df = df[['1Wk','1M','3M','6M','1Yr']] * 100

    return df.iloc[-1].to_string()


def get_seasonality (ticker):

    # Calculate the start date as the first day of the month, 10 years ago from today
    today = datetime.now()
    start_date = datetime(today.year - 10, today.month, 1)

    # Use yfinance to fetch data
    data = yf.download(ticker, start=start_date.strftime('%Y-%m-%d'), end=today.strftime('%Y-%m-%d'))

    # Resample the data to monthly frequency, calculating the monthly returns
    monthly_returns = data['Adj Close'].resample('M').ffill().pct_change()

    # Reset index to get date components
    monthly_returns = monthly_returns.reset_index()

    # Extract year and month from the date
    monthly_returns['year'] = monthly_returns['Date'].dt.year
    monthly_returns['month'] = monthly_returns['Date'].dt.month

    # Create a pivot table with years as rows and months as columns
    monthly_returns_pivot = monthly_returns.pivot(index='year', columns='month', values='Adj Close')

    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

    # Rename columns for clarity
    monthly_returns_pivot.columns = months #[f'Month {col}' for col in monthly_returns_pivot.columns]

    # Calculate metrics
    average_returns = monthly_returns_pivot.mean()
    std_dev = monthly_returns_pivot.std()
    ratio = average_returns / std_dev

    max_returns = monthly_returns_pivot.max()
    min_returns = monthly_returns_pivot.min()

    # Calculate the hit ratio
    positive_returns = monthly_returns_pivot > 0
    hit_ratio = monthly_returns_pivot[positive_returns].count() / monthly_returns_pivot.count()

    # Construct the table
    table = pd.DataFrame({
        'Month': months,
        'Average Returns (A)': average_returns.values,
        'Standard Deviation (B)': std_dev.values,
        'A/B': ratio.values,
        'Max Returns': max_returns.values,
        'Min Returns': min_returns.values,
        'Hit Ratio': hit_ratio.values
    })

    # Set Month as the index
    table.set_index('Month', inplace=True)

    return table


def get_daily_seasonality(ticker):
    # Calculate the start date as the first day of the month, 10 years ago from today
    today = datetime.now()
    start_date = datetime(today.year - 10, 1, 1)

    # Use yfinance to fetch data
    data = yf.download(ticker, start=start_date.strftime('%Y-%m-%d'), end=today.strftime('%Y-%m-%d'))

    # Create a complete index for each year to ensure all DayOfYear values are present
    full_range = pd.date_range(start=start_date, end=today, freq='D')
    full_data = pd.DataFrame(index=full_range).join(data['Adj Close'], how='left')

    # Forward fill missing data
    full_data['Adj Close'] = full_data['Adj Close'].ffill()

    # Create 'Year' and 'DayOfYear' columns
    full_data['Year'] = full_data.index.year
    full_data['DayOfYear'] = full_data.index.dayofyear

    # Remove rows where DayOfYear is greater than 365
    full_data = full_data[full_data['DayOfYear'] <= 365]

    # Normalize each year so that the first day of the year is 100
    full_data['Normalized'] = full_data.groupby('Year')['Adj Close'].transform(lambda x: (x / x.iloc[0]) * 100)

    # Get the current year
    current_year = today.year

    # Calculate the average normalized price for the last 10 years excluding the current year
    avg_normalized = full_data[full_data['Year'] != current_year].groupby('DayOfYear')['Normalized'].mean()

    # Plotting the daily seasonality chart
    plt.figure(figsize=(8, 5))
    for year in full_data['Year'].unique():
        yearly_data = full_data[full_data['Year'] == year]
        if year == current_year:
            plt.plot(yearly_data['DayOfYear'], yearly_data['Normalized'], label=str(year), color='#ff6600')
        else:
            plt.plot(yearly_data['DayOfYear'], yearly_data['Normalized'], label=str(year), color='gray', alpha=0.5)

    # Plot the average normalized price for the last 10 years
    plt.plot(avg_normalized.index, avg_normalized.values, label='10-Year Average', color='#0066cc', linewidth=2)

    plt.xlabel('Day of Year')
    plt.ylabel('Normalized Price (Start of Year = 100)')
    plt.title(f'Daily Seasonality Chart for {ticker}')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.grid(True)
    plt.show()


def get_analysis(ticker):
    headers = {"user-agent": "Mozilla/5.0"}
    res = requests.get(f"https://finance.yahoo.com/quote/{ticker}/analysis?p={ticker}", headers=headers).text

    # 0 - Earnings Estimate, 1 - Revenue Estimate, 2 - Earnings History, 3 - EPS Trend, 4 - EPS Revision, 5 - Growth Estimates
    res = pd.read_html(res)

    df1 = res[2]
    df2 = res[0]
    eps_revision = res[4].replace('--', 0)

    # 모든 값을 숫자로 변환
    for col in eps_revision.columns[1:]:
        eps_revision[col] = pd.to_numeric(eps_revision[col])

    # 데이터 정리
    eps_est_actual_idx = df1['CURRENCY IN USD'].isin(['EPS Est.', 'EPS Actual'])
    eps_est_actual = df1.iloc[eps_est_actual_idx.values, 1:].transpose()
    eps_est_actual.columns = ['EPS Est.', 'EPS Actual']
    for col in eps_est_actual.columns:
        eps_est_actual[col] = pd.to_numeric(eps_est_actual[col].str.replace('%', ''), errors='coerce')

    # 과거 4분기의 마지막 날짜를 기준으로 향후 2분기의 날짜 계산
    last_date = pd.to_datetime(eps_est_actual.index[-1])

    future_dates = [(last_date + pd.DateOffset(months=3)).strftime('%m/%d/%Y'),
                    (last_date + pd.DateOffset(months=6)).strftime('%m/%d/%Y')]

    # future_eps_est는 향후 2분기의 EPS 평균 추정치, 최저 추정치, 최고 추정치를 포함
    future_eps_est_idx = df2['CURRENCY IN USD'].isin(['Avg. Estimate', 'Low Estimate', 'High Estimate'])
    future_eps_est = df2.iloc[future_eps_est_idx.values, [1, 2]].transpose()
    future_eps_est.columns = ['EPS Est.', 'Low Estimate', 'High Estimate']
    future_eps_est.index = future_dates

    eps_est = pd.concat([eps_est_actual, future_eps_est])
    eps_est['Surprise'] = (eps_est['EPS Actual'] - eps_est['EPS Est.'])

    # Surprise 값을 이동 평균으로 보정
    window_size = 2  # 이동 평균 윈도우 크기
    eps_est['Surprise Smoothed'] = eps_est['Surprise'].rolling(window=window_size, min_periods=1).mean()

    # 차트 그리기
    plt.figure(figsize=(8, 5))

    # 지난 4분기 EPS 실제 발표치와 과거 4분기 추정치
    plt.plot(eps_est.index, eps_est['EPS Est.'], marker='o', label='EPS Est', linestyle='-', color='#005BAC')  # 미래에셋 블루
    plt.errorbar(eps_est.index, eps_est['EPS Actual'], fmt='o', color='#FF8C00', label='EPS Actual')  # 미래에셋 오렌지

    # 향후 2분기 EPS 최저/최고 범위
    plt.fill_between(future_eps_est.index, future_eps_est['Low Estimate'], future_eps_est['High Estimate'], color='gray', alpha=0.2, label='EPS Range (Low - High)')

    # EPS 수치와 Surprise 표시
    for i, row in eps_est.iterrows():
        if not pd.isna(row['EPS Actual']):
            plt.text(i, row['EPS Actual'], f"{row['EPS Actual']:.2f}\n({row['Surprise']:.2f})", ha='center', va='bottom', fontsize=11, fontweight='bold', color='#FF8C00')
        if not pd.isna(row['EPS Est.']):
            plt.text(i, row['EPS Est.'], f"{row['EPS Est.']:.2f}", ha='center', va='top', fontsize=11, fontweight='bold', color='#005BAC')

    # EPS Revision 화살표 표시
    def add_arrow(ax, x, y, direction):
        dx = 0
        dy = np.abs(y)*0.05 if direction == 'up' else -np.abs(y*0.05)
        color = 'green' if direction == 'up' else 'red'
        ax.arrow(x, y+dy, dx, dy, width = 0.05, head_width=0.1, head_length=dy, fc=color, ec=color)

    ax = plt.gca()

    for i, future_date in enumerate(future_dates):
        up_last_7_days = eps_revision.iloc[0, i + 1]
        down_last_7_days = eps_revision.iloc[2, i + 1]
        up_last_30_days = eps_revision.iloc[1, i + 1]
        down_last_30_days = eps_revision.iloc[3, i + 1]

        diff_last_7_days = up_last_7_days - down_last_7_days
        diff_last_30_days = up_last_30_days - down_last_30_days

        if diff_last_7_days > 0:
            add_arrow(ax, i + len(eps_est_actual)-0.05, future_eps_est.iloc[i]['EPS Est.'], 'up')
        elif diff_last_7_days < 0:
            add_arrow(ax, i + len(eps_est_actual)-0.05, future_eps_est.iloc[i]['EPS Est.'], 'down')
        if diff_last_30_days > 0:
            add_arrow(ax, i + len(eps_est_actual)+0.05, future_eps_est.iloc[i]['EPS Est.'], 'up')
        elif diff_last_30_days < 0:
            add_arrow(ax, i + len(eps_est_actual)+0.05, future_eps_est.iloc[i]['EPS Est.'], 'down')

    plt.xlabel('Quarters')
    plt.ylabel('EPS')
    plt.title(f'{ticker} EPS Estimates and Actuals', fontsize=14, fontweight='bold')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(rotation=45, fontsize=11)
    plt.yticks(fontsize=11)
    plt.tight_layout()
    plt.show()

    return eps_est


def get_analyst_target(ticker):
    company = yf.Ticker(ticker)
    info = company.info

    # 예외 처리 추가
    current_price = info.get('currentPrice', None)
    if current_price is None:
        raise ValueError("현재 가격 정보를 가져올 수 없습니다.")

    target_mean_price = info.get('targetMeanPrice', None)
    target_low_price = info.get('targetLowPrice', None)
    target_high_price = info.get('targetHighPrice', None)
    number_of_analysts = info.get('numberOfAnalystOpinions', None)

    # 기대수익률 계산
    expected_return_mean = ((target_mean_price - current_price) / current_price) * 100 if target_mean_price else None
    expected_return_low = ((target_low_price - current_price) / current_price) * 100 if target_low_price else None
    expected_return_high = ((target_high_price - current_price) / current_price) * 100 if target_high_price else None

    # 데이터 다운로드 (5년간의 일일 데이터로 변경)
    data = yf.download(ticker, period='5y', interval='1d')

    if data.empty:
        raise ValueError("주식 데이터를 가져올 수 없습니다.")

    # 1년 후 날짜 추가
    future_date = data.index[-1] + relativedelta(years=1)

    # 차트 생성
    plt.figure(figsize=(8, 5))
    plt.plot(data.index, data['Adj Close'], label='Historical Price', color='#1E90FF')  # 블루 색상
    plt.scatter(data.index[-1], current_price, color='#FF8C00', zorder=5)  # 오렌지 색상

    if target_low_price and target_high_price and target_mean_price:
        # 목표가 범위와 현재가 연결
        plt.plot([data.index[-1], future_date], [current_price, target_low_price], color='green', linestyle='--')
        plt.plot([data.index[-1], future_date], [current_price, target_high_price], color='red', linestyle='--')
        plt.plot([data.index[-1], future_date], [current_price, target_mean_price], color='#FF8C00', linestyle='--')  # 오렌지 색상

        # 목표가 범위 표시
        #plt.fill_between([data.index[-1], future_date], current_price, target_low_price, color='gray', alpha=0.3)
        #plt.fill_between([data.index[-1], future_date], current_price, target_high_price, color='gray', alpha=0.3, label='Target Range')

        # 차트 상에 값 표시
        plt.text(future_date, target_low_price, f'Target Min: {target_low_price} ({expected_return_low:.2f}%)', color='green', fontsize=10, verticalalignment='bottom')
        plt.text(future_date, target_high_price, f'Target Max: {target_high_price} ({expected_return_high:.2f}%)', color='red', fontsize=10, verticalalignment='bottom')
        plt.text(future_date, target_mean_price, f'Target Mean: {target_mean_price} ({expected_return_mean:.2f}%)', color='#FF8C00', fontsize=10, verticalalignment='bottom')

    # 현재가 표시
    plt.text(data.index[-1], current_price, f'Current Price: {current_price}', color='#FF8C00', fontsize=10, verticalalignment='bottom')

    # 차트 구성
    plt.title(f'5-Year Stock Price Movement and 1-Year Target Prices for {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()

    plt.show()


def get_analyst_recommendations(ticker):
    company = yf.Ticker(ticker)

    recommendations_df = company.recommendations

    recommendations_df_normalized = recommendations_df.set_index('period').apply(lambda x: x / x.sum() * 100, axis=1)

    # Plotting
    fig, ax = plt.subplots(figsize=(10, 6))

    # Define the bottom for the stacking
    bottom = pd.Series([0] * len(recommendations_df_normalized), index=recommendations_df_normalized.index)

    # Colors for different ratings using Mirae Asset colors
    colors = ['#F58220', '#043B72', '#F0B26B', '#CB6015', '#AD624E', '#84888B', '#00A9CE', '#0086B8']

    # Plot each rating as a bar
    for i, column in enumerate(recommendations_df_normalized.columns):
        ax.bar(recommendations_df_normalized.index, recommendations_df_normalized[column], bottom=bottom, label=column, color=colors[i % len(colors)])
        bottom += recommendations_df_normalized[column]

    # Add the values to the bars
    for idx in recommendations_df_normalized.index:
        cumulative = 0
        for col in recommendations_df_normalized.columns:
            height = recommendations_df_normalized.loc[idx, col]
            value = recommendations_df.loc[recommendations_df["period"] == idx, col].values[0]
            ax.text(idx, cumulative + height / 2, f'{value:.2f}', ha='center', va='center', fontsize=8, color='black')
            cumulative += height

    # Adding title and labels
    plt.title('Financial Expert Ratings Over Time (Percentage)', fontsize=14, fontweight='bold')
    plt.xlabel('Period', fontsize=12)
    plt.ylabel('Percentage', fontsize=12)
    plt.legend(loc='upper left', title='Ratings')
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)

    # Improving layout
    plt.xticks(rotation=45)
    plt.tight_layout()

    plt.show()


def get_analyst_upgrade (ticker):

    company = yf.Ticker(ticker)

    # Get the upgrades and downgrades data
    upgrades_downgrades = company.upgrades_downgrades

    # Ensure the 'GradeDate' column is in the DataFrame
    if 'GradeDate' not in upgrades_downgrades.columns:
        upgrades_downgrades.reset_index(inplace=True)

    # Create a new column 'Action_Value' based on the 'Action' column
    upgrades_downgrades['Action_Value'] = upgrades_downgrades['Action'].apply(lambda x: 1 if x.lower() == 'up' else -1 if x.lower() == 'down' else 0)

    # Convert the 'GradeDate' column to datetime
    upgrades_downgrades['GradeDate'] = pd.to_datetime(upgrades_downgrades['GradeDate'])

    # Aggregate the data by date, summing up the action values
    daily_action_values = upgrades_downgrades.groupby(upgrades_downgrades['GradeDate'].dt.date)['Action_Value'].sum().reset_index()
    daily_action_values.rename(columns={'GradeDate': 'Date'}, inplace=True)

    # Set 'Date' as the index
    daily_action_values.set_index('Date', inplace=True)

    # Calculate the cumulative sum of action values
    daily_action_values['Cumulative_Value'] = daily_action_values['Action_Value'].rolling(21).sum()

    # Get the start date for fetching stock price data
    start_date = daily_action_values.index.min().strftime('%Y-%m-%d')

    # Fetch the stock price data starting from the earliest date in the upgrades and downgrades data
    stock_data = yf.download(ticker, start=start_date)

    # Merge the daily action values data with the stock price data
    merged_data = pd.merge(daily_action_values, stock_data['Adj Close'], left_index=True, right_index=True, how='left')

    # Plotting the data with two different y-axes
    fig, ax1 = plt.subplots(figsize=(10, 5))

    color = 'tab:blue'
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Cumulative Value', color=color)
    ax1.fill_between(merged_data.index, 0, merged_data['Cumulative_Value'], color=color, alpha=0.5)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis
    color = 'tab:red'
    ax2.set_ylabel('Adj Close', color=color)  # We already handled the x-label with ax1
    ax2.plot(merged_data.index, merged_data['Adj Close'], color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # Otherwise, the right y-label is slightly clipped
    plt.title(f'{ticker} Cumulative Upgrades/Downgrades vs. Adjusted Close Price')
    plt.show()


def get_financial_statements(ticker):

    """
    if "." in ticker:
        ticker=ticker.split(".")[0]
    else:
        ticker=ticker
    """

    company = yf.Ticker(ticker)
    balance_sheet = company.balance_sheet
    if balance_sheet.shape[1]>=3:
        balance_sheet=balance_sheet.iloc[:,:3]    # Remove 4th years data
    balance_sheet=balance_sheet.dropna(how="any")
    balance_sheet = balance_sheet.to_string()
    return balance_sheet


def get_income_statements(ticker):
    stock = yf.Ticker(ticker)
    quarterly_income_statement = stock.quarterly_financials
    #quarterly_income_statement = stock.financials

    # Transpose the DataFrame for easier processing
    quarterly_income_statement = quarterly_income_statement.T

    # Extract relevant fields for the most recent quarter
    latest_quarter = quarterly_income_statement.index[0]
    income_statement_data = {
        'Total Revenue': quarterly_income_statement['Total Revenue'][latest_quarter],
        'Cost of Revenue': quarterly_income_statement['Cost Of Revenue'][latest_quarter],
        'Gross Profit': quarterly_income_statement['Gross Profit'][latest_quarter],
        'Operating Expenses': quarterly_income_statement['Operating Expense'][latest_quarter],
        'Research and Development': quarterly_income_statement['Research And Development'][latest_quarter],
        'Selling General and Administrative': quarterly_income_statement['Selling General And Administration'][latest_quarter],
        'Operating Income': quarterly_income_statement['Operating Income'][latest_quarter],
        'Tax Provision': quarterly_income_statement['Tax Provision'][latest_quarter],
        'Net Income': quarterly_income_statement['Net Income'][latest_quarter]
    }

    # Define the source and target nodes for the Sankey diagram
    sources = ['Total Revenue', 'Total Revenue', 'Gross Profit', 'Gross Profit', 'Operating Income', 'Operating Income', 'Operating Expenses', 'Operating Expenses']
    targets = ['Cost of Revenue', 'Gross Profit', 'Operating Income', 'Operating Expenses', 'Net Income', 'Tax Provision', 'Research and Development', 'Selling General and Administrative']
    values = [income_statement_data['Cost of Revenue'],
              income_statement_data['Gross Profit'],
              income_statement_data['Operating Income'],
              income_statement_data['Operating Expenses'],
              income_statement_data['Net Income'],
              income_statement_data['Tax Provision'],
              income_statement_data['Research and Development'],
              income_statement_data['Selling General and Administrative']]

    # Create a list of all nodes
    all_nodes = list(set(sources + targets))

    # Map nodes to indices
    node_indices = {node: idx for idx, node in enumerate(all_nodes)}

    # Convert source and target labels to indices
    source_indices = [node_indices[src] for src in sources]
    target_indices = [node_indices[tgt] for tgt in targets]

    # Define colors for the nodes for better visualization
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']

    # Create the Sankey diagram
    fig = go.Figure(data=[go.Sankey(
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=[f"{node}: ${income_statement_data[node]:,.0f}" if node in income_statement_data else node for node in all_nodes],
            color=colors[:len(all_nodes)]
        ),
        link=dict(
            source=source_indices,
            target=target_indices,
            value=values,
            color='rgba(0,0,0,0.4)'  # Semi-transparent black for links
        ))])

    # Set chart title and update layout for better readability
    fig.update_layout(
        title_text=f"{ticker} Income Statement Sankey Diagram (Latest Quarter)",
        font_size=12,
        title_font_size=14,
        title_x=0.5,
    )

    # Show the figure
    fig.show()


def earnings_post_pattern(ticker, years=10, method='winsorize'):

    # 주식 데이터 가져오기
    stock = yf.Ticker(ticker)
    hist = stock.history(period=f"{years}y")

    # 실적 데이터 가져오기
    earnings = stock.get_earnings_dates(limit=years * 4)
    earnings.drop_duplicates(subset=None, keep='first', inplace=True)

    # 날짜 형식 맞추기
    earnings.index = pd.to_datetime(earnings.index.strftime('%Y-%m-%d'))
    hist.index = pd.to_datetime(hist.index.strftime('%Y-%m-%d'))

    # 데이터 병합
    merged_df = pd.concat([earnings, hist], axis=1).sort_index(ascending=False)

    # 데이터 처리
    merged_df['Close'].fillna(method='ffill', inplace=True)
    merged_df['ret'] = merged_df['Close'].pct_change(20).shift(-20)*100

    # 필터링된 데이터
    filtered_df = merged_df.loc[earnings.index]
    filtered_df = filtered_df[['EPS Estimate', 'Reported EPS', 'Surprise(%)', 'ret']]

    # 3σ 기준을 넘는 데이터 처리
    #z_scores = zscore(filtered_df[['Surprise(%)', 'ret']])
    #if method == 'trimming':
    #    filtered_df = filtered_df[(np.abs(z_scores) < 3).all(axis=1)]
    #elif method == 'winsorize':
    #    filtered_df[['Surprise(%)', 'ret']] = np.where(np.abs(z_scores) > 3, np.sign(z_scores) * 3, filtered_df[['Surprise(%)', 'ret']])

    return filtered_df.to_string()


def earnings_timeseries_pattern(ticker):

    def get_earnings_dates(ticker):
        # 주식 데이터 가져오기
        stock = yf.Ticker(ticker)
        years = 10

        # 실적 데이터 가져오기
        earnings = stock.get_earnings_dates(limit=years*4)
        earnings.drop_duplicates(subset=None, keep='first', inplace=True)

        # 날짜 형식 맞추기 및 리스트로 변환
        earnings_dates = list(earnings.index.strftime('%Y-%m-%d'))

        return earnings_dates


    def find_next_business_day(release_date_list):
        # 현재 날짜 가져오기
        today = datetime.now().date()

        # 문자열 형식의 날짜 리스트를 datetime.date 객체로 변환
        dates = [datetime.strptime(date, '%Y-%m-%d').date() for date in release_date_list]

        # 오늘 날짜보다 크거나 같은 날짜 찾기
        future_dates = [date for date in dates if date >= today]

        # 오늘 날짜보다 가까운 미래 날짜 찾기
        if future_dates:
            next_business_day = min(future_dates)
            # 남은 영업일 수 계산
            remaining_business_days = np.busday_count(today, next_business_day)
            return next_business_day, remaining_business_days
        else:
            return None, 0

    def plot_normalized_performance(release_date_list, ticker):

        wnd_length = 21
        current_date = datetime.now()
        data = yf.download(ticker, start=datetime(1990, 1, 1), end=current_date)

        samples = []
        earnings_dates = []
        for target_date in release_date_list:
            target_date = pd.to_datetime(target_date)
            if target_date in data.index:
                earnings_dates.append(target_date)
                target_index = data.index.get_loc(target_date)
                start_index = target_index - wnd_length
                end_index = target_index + wnd_length
                selected_data = data.iloc[start_index:end_index + 1]['Adj Close'] / data.iloc[target_index]['Adj Close'] * 100
                samples.append(selected_data.values)

        # 두 번째 차트: 최근 2년 및 10년 동안의 평균 성과

        # "time to earnings release"를 표시하는 문자열 리스트 생성
        index_labels = ['Time To Earnings Release: ' + str(i) for i in range(-21, 22)]

        #result = pd.DataFrame(samples, columns=range(-wnd_length, wnd_length + 1), index=earnings_dates )
        result = pd.DataFrame(samples, columns=index_labels, index=earnings_dates )
        return result.mean().to_string()

    earnings_dates = get_earnings_dates(ticker)

    return plot_normalized_performance(earnings_dates, ticker)


sample_ticker = 'NVDA'

#get_stock_performance (sample_ticker)
seasonality_table = get_seasonality (sample_ticker)
get_daily_seasonality (sample_ticker)
get_analyst_target(sample_ticker)
get_analyst_recommendations (sample_ticker)
get_analyst_upgrade (sample_ticker)
get_analysis (sample_ticker)

print (seasonality_table)
# Figure와 Axes 객체 생성
fig, ax1 = plt.subplots()

# bar 차트 그리기
ax1.bar(seasonality_table['Average Returns (A)'].index, seasonality_table['Average Returns (A)'], color='b', alpha=0.5)
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Average Returns (A)', color='b')

# 두 번째 y 축을 위한 ax2 객체 생성
ax2 = ax1.twinx()

# 선 그래프 그리기
ax2.plot(seasonality_table['A/B'].index, seasonality_table['A/B'], color='r', marker='o')
ax2.set_ylabel('Returns Avg/Std', color='r')

# 그래프 제목 설정
plt.title('Average Returns (A) and A/B Comparison')

# 그래프 보여주기
plt.show()
#print (get_financial_statements (sample_ticker))
#get_income_statements (sample_ticker)
#earnings_post_pattern (sample_ticker)
#earnings_timeseries_pattern (sample_ticker)
```

```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
table = soup.find('table', {'id': 'constituents'})

tickers = []
for row in table.findAll('tr')[1:]:
    ticker = row.findAll('td')[0].text.strip()
    tickers.append(ticker)
```

```python
data = []

for ticker in tickers:
    try:
        stock = yf.Ticker(ticker)
        info = stock.info

        #for k,v in stock.info.items():
        #    print (k, '\t', v, '\n')

        current_price = info['currentPrice']
        target_mean_price = info.get('targetMeanPrice')
        target_low_price = info.get('targetLowPrice')
        number_of_analysts = info.get('numberOfAnalystOpinions')

        if current_price and target_mean_price and target_low_price and number_of_analysts and number_of_analysts >= 5:
            expected_return = (target_mean_price - current_price) / current_price * 100
            worst_case_return = (target_low_price - current_price) / current_price * 100

            data.append({
                'Ticker': ticker,
                'Industry': info.get('industry', 'Unknown'),
                'Sector': info.get('Sector', 'Unknown'),
                'Current Price': current_price,
                'Target Mean Price': target_mean_price,
                'Expected Return (%)': expected_return,
                'Target Low Price': target_low_price,
                'Worst Case Return (%)': worst_case_return,
                'Number of Analysts': number_of_analysts
            })
    except Exception as e:
        print(f"Error processing ticker {ticker}: {e}")

df = pd.DataFrame(data)
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

target_industry = "Semiconductors"

# 산업에 속한 종목 리스트를 가져오기
industry_tickers = df[df["Industry"] == target_industry]['Ticker'].values

# 종목별 실적 데이터를 저장할 리스트 초기화
all_earnings = []

# 각 종목의 실적 데이터를 가져오기
for ticker in industry_tickers:
    stock = yf.Ticker(ticker)
    earnings = stock.earnings_dates.reset_index()
    earnings['Ticker'] = ticker
    all_earnings.append(earnings)

# 모든 종목의 실적 데이터를 하나의 데이터프레임으로 병합
all_earnings_df = pd.concat(all_earnings)

all_earnings_df['Surprise(%)'] = (all_earnings_df['Reported EPS'] - all_earnings_df['EPS Estimate']) / np.abs(all_earnings_df['EPS Estimate']) * 100

# 결측치 제거
all_earnings_df.dropna(subset=['Surprise(%)'], inplace=True)

# 날짜별로 정렬
all_earnings_df['Earnings Date'] = pd.to_datetime(all_earnings_df['Earnings Date']).dt.tz_localize(None)
all_earnings_df.sort_values(by='Earnings Date', inplace=True)

# 최근 1년간의 데이터 필터링
one_year_ago = pd.Timestamp.now() - pd.DateOffset(years=1)
one_year_ago = one_year_ago.tz_localize(None)
recent_earnings_df = all_earnings_df[all_earnings_df['Earnings Date'] >= one_year_ago]

# 이상치 처리 (IQR 방법)
Q1 = recent_earnings_df['Surprise(%)'].quantile(0.25)
Q3 = recent_earnings_df['Surprise(%)'].quantile(0.75)
IQR = Q3 - Q1

# 이상치 범위 설정
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 이상치 제거
filtered_earnings_df = recent_earnings_df[(recent_earnings_df['Surprise(%)'] >= lower_bound) &
                                          (recent_earnings_df['Surprise(%)'] <= upper_bound)]



# 중복 제거: 날짜를 제외한 중복된 행을 찾고, 가장 빠른 날짜만 남기고 나머지를 제거
#filtered_earnings_df = filtered_earnings_df.sort_values(by='Earnings Date')
filtered_earnings_df = filtered_earnings_df.drop_duplicates(subset=['EPS Estimate', 'Reported EPS', 'Surprise(%)', 'Ticker'], keep='first')

# Rolling 계산을 위해 날짜를 인덱스로 설정
filtered_earnings_df.set_index('Earnings Date', inplace=True)

# Surprise 평균과 표준 편차 계산
st_wnd = len(filtered_earnings_df['Ticker'].unique())

surp_mean = filtered_earnings_df['Surprise(%)'].rolling(window=st_wnd).mean()
surp_std = filtered_earnings_df['Surprise(%)'].rolling(window=st_wnd).std()

# 두 번 이동 평균 적용하여 부드럽게
smooth_surp_mean = surp_mean.rolling(window=st_wnd).mean()
smooth_surp_std = surp_std.rolling(window=st_wnd).mean()

# 차트 생성
plt.figure(figsize=(14, 7))
for ticker in industry_tickers:
    ticker_data = filtered_earnings_df[filtered_earnings_df['Ticker'] == ticker]
    plt.plot(ticker_data.index, ticker_data['Surprise(%)'], marker='o', linestyle='-', label=f'{ticker} Surprise(%)')

# Rolling 평균과 표준 편차 추가 (부드럽게)
plt.plot(smooth_surp_mean, color='blue', linestyle='--', label='Smoothed Surprise Mean')
plt.fill_between(smooth_surp_std.index, smooth_surp_mean - smooth_surp_std, smooth_surp_mean + smooth_surp_std, color='blue', alpha=0.1, label='Smoothed Surprise Rolling Std')

plt.xlabel('Earnings Date')
plt.ylabel('Surprise(%)')
plt.title('Earnings Surprise by Date')
plt.legend()
plt.grid(True)
plt.show()
```

#옵션 만기별 행사 확률 계산

```python
import numpy as np
import pandas as pd
import yfinance as yf
import scipy.stats as si
import datetime
import matplotlib.pyplot as plt
from dateutil.relativedelta import relativedelta
import pdb


def calculate_itm_probability(S, K, T, r, sigma, option_type='call'):
    if sigma == 0:
        return 0
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == 'call':
        return si.norm.cdf(d2)
    elif option_type == 'put':
        return si.norm.cdf(-d2)


def plot_itm_probability(ax, ticker, target_prices, ylim):

    target_mean_price, target_low_price, target_high_price = target_prices

    stock = yf.Ticker(ticker)
    expirations = stock.options

    current_price = stock.history(period='1d')['Close'].iloc[-1]
    risk_free_rate = 0.05

    closest_expiration = min(expirations, key=lambda x: abs((datetime.datetime.strptime(x, '%Y-%m-%d').date() - (datetime.date.today() + datetime.timedelta(days=365)))))

    opt = stock.option_chain(closest_expiration)
    calls = opt.calls.assign(Expiration=closest_expiration, Type='Call')
    puts = opt.puts.assign(Expiration=closest_expiration, Type='Put')

    expiration_date = datetime.datetime.strptime(closest_expiration, '%Y-%m-%d').date()
    days_to_expiration = (expiration_date - datetime.date.today()).days / 365.0

    calls['ITM_Probability'] = calls.apply(lambda row: calculate_itm_probability(
        current_price, row['strike'], days_to_expiration, risk_free_rate, row['impliedVolatility'], 'call'), axis=1)
    puts['ITM_Probability'] = puts.apply(lambda row: calculate_itm_probability(
        current_price, row['strike'], days_to_expiration, risk_free_rate, row['impliedVolatility'], 'put'), axis=1)

    calls_for_upside = calls[calls['strike'] >= current_price]
    puts_for_downside = puts[puts['strike'] < current_price]

    options_data = pd.concat([puts_for_downside, calls_for_upside])
    options_data = options_data.sort_values(by='strike')
    options_data = options_data.dropna(subset=['ITM_Probability'])

    itm_probabilities = options_data['ITM_Probability'].values
    for i in range(len(itm_probabilities)):
        if itm_probabilities[i] == 0:
            left_indices = range(max(0, i - 3), i)
            right_indices = range(i + 1, min(len(itm_probabilities), i + 4))
            neighbor_indices = list(left_indices) + list(right_indices)
            if neighbor_indices:
                itm_probabilities[i] = np.mean([itm_probabilities[j] for j in neighbor_indices if itm_probabilities[j] != 0])

    options_data['ITM_Probability'] = itm_probabilities
    options_data['SMA_ITM_Probability'] = options_data['ITM_Probability'].rolling(window=11, center=True).mean()
    options_data = options_data.dropna(subset=['SMA_ITM_Probability'])

    pdb.set_trace()

    ax.plot(options_data['SMA_ITM_Probability'], options_data['strike'], label='Smoothed Probability')
    ax.set_xlabel('Probability')
    ax.set_ylabel('Price')
    ax.set_title(f'Probability of [{ticker}] Target Price on Option Market')
    ax.set_ylim(ylim)

    if target_mean_price:
        ax.axhline(y=target_mean_price, color='#FF8C00', linestyle='--', label='Target Mean')
    if target_low_price:
        ax.axhline(y=target_low_price, color='green', linestyle='--', label='Target Min')
    if target_high_price:
        ax.axhline(y=target_high_price, color='red', linestyle='--', label='Target Max')

    up_20 = current_price * (1 + 0.20)
    down_20 = current_price * (1 - 0.20)
    ax.axhline(y=up_20, color='gray', linestyle=':', label='20% Up')
    ax.axhline(y=down_20, color='gray', linestyle=':', label='20% Down')

    def get_nearest_probability(target_price):
        nearest_strike = options_data.iloc[(options_data['strike'] - target_price).abs().argsort()[:1]]
        return nearest_strike['SMA_ITM_Probability'].values[0] if not nearest_strike.empty else None

    target_prob_mean = get_nearest_probability(target_mean_price)
    target_prob_low = get_nearest_probability(target_low_price)
    target_prob_high = get_nearest_probability(target_high_price)

    up_20_prob = get_nearest_probability(up_20)
    down_20_prob = get_nearest_probability(down_20)

    """
    if target_prob_mean:
        ax.text(target_prob_mean, target_mean_price, f'{target_prob_mean:.2%}', color='#FF8C00', fontsize=10, verticalalignment='bottom')
    if target_prob_low:
        ax.text(target_prob_low, target_low_price, f'{target_prob_low:.2%}', color='green', fontsize=10, verticalalignment='bottom')
    if target_prob_high:
        ax.text(target_prob_high, target_high_price, f'{target_prob_high:.2%}', color='red', fontsize=10, verticalalignment='bottom')
    """

    if up_20_prob:
        ax.text(up_20_prob, up_20, f'{up_20_prob:.2%}', color='black', fontsize=10, verticalalignment='bottom')
    if down_20_prob:
        ax.text(down_20_prob, down_20, f'{down_20_prob:.2%}', color='black', fontsize=10, verticalalignment='bottom')

    ax.legend()


# Analyst Target Chart
def get_analyst_target(ax, ticker):
    company = yf.Ticker(ticker)
    info = company.info

    current_price = info.get('currentPrice', None)
    if current_price is None:
        raise ValueError("현재 가격 정보를 가져올 수 없습니다.")

    target_mean_price = info.get('targetMeanPrice', None)
    target_low_price = info.get('targetLowPrice', None)
    target_high_price = info.get('targetHighPrice', None)
    number_of_analysts = info.get('numberOfAnalystOpinions', None)

    expected_return_mean = ((target_mean_price - current_price) / current_price) * 100 if target_mean_price else None
    expected_return_low = ((target_low_price - current_price) / current_price) * 100 if target_low_price else None
    expected_return_high = ((target_high_price - current_price) / current_price) * 100 if target_high_price else None

    data = yf.download(ticker, period='5y', interval='1d')
    if data.empty:
        raise ValueError("주식 데이터를 가져올 수 없습니다.")

    future_date = data.index[-1] + relativedelta(years=1)

    ax.plot(data.index, data['Adj Close'], label='Historical Price', color='#1E90FF')
    ax.scatter(data.index[-1], current_price, color='#FF8C00', zorder=5)

    if target_low_price and target_high_price and target_mean_price:
        ax.plot([data.index[-1], future_date], [current_price, target_low_price], color='green', linestyle='--')
        ax.plot([data.index[-1], future_date], [current_price, target_high_price], color='red', linestyle='--')
        ax.plot([data.index[-1], future_date], [current_price, target_mean_price], color='#FF8C00', linestyle='--')

        ax.text(future_date, target_low_price, f'Target Min: {target_low_price} ({expected_return_low:.2f}%)', color='green', fontsize=10, verticalalignment='bottom')
        ax.text(future_date, target_high_price, f'Target Max: {target_high_price} ({expected_return_high:.2f}%)', color='red', fontsize=10, verticalalignment='bottom')
        ax.text(future_date, target_mean_price, f'Target Mean: {target_mean_price} ({expected_return_mean:.2f}%)', color='#FF8C00', fontsize=10, verticalalignment='bottom')

    #ax.text(data.index[-1], current_price, f'Current Price: {current_price}', color='#FF8C00', fontsize=10, verticalalignment='bottom')

    ax.set_title(f'5-Year Stock Price Movement and 1-Year Target Prices for {ticker}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    ax.legend()
    ylim = ax.get_ylim()
    target_prices = (target_mean_price, target_low_price, target_high_price)

    return target_prices, ylim

# Combined Plot
def plot_combined(ticker):
    fig, axes = plt.subplots(1, 2, figsize=(16, 7))

    # Plot Analyst Target
    target_prices, ylim = get_analyst_target(axes[0], ticker)

    # Plot ITM Probability
    plot_itm_probability(axes[1], ticker, target_prices, ylim)

    plt.tight_layout()
    plt.show()

# Example usage
plot_combined('AAPL')
```
